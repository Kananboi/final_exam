```python
from abc import ABC, abstractmethod
class Kaynak(ABC):
    """Tüm kaynakların (Kitap, Dergi) ortak özelliklerini barındıran soyut sınıf."""

    def __init__(self, baslik, kayit_no):
        self._baslik = baslik
        self._kayit_no = kayit_no

    @property
    def baslik(self):
        return self._baslik

    @baslik.setter
    def baslik(self, deger):
        if not deger:
            raise ValueError("Başlık boş olamaz.")
        self._baslik = deger

    @property
    def kayit_no(self):
        return self._kayit_no

    @kayit_no.setter
    def kayit_no(self, deger):
        if not deger:
            raise ValueError("Kayıt numarası boş olamaz.")
        self._kayit_no = deger

    @abstractmethod
    def __str__(self):
        pass

class Kitap(Kaynak):
    def __init__(self, baslik, kayit_no, yazar, sayfa_sayisi):
        super().__init__(baslik, kayit_no)
        self._yazar = yazar
        self._sayfa_sayisi = sayfa_sayisi

    @property
    def yazar(self):
        return self._yazar

    @yazar.setter
    def yazar(self, deger):
        if not deger:
            raise ValueError("Yazar boş olamaz.")
        self._yazar = deger
    @property
    def sayfa_sayisi(self):
        return self._sayfa_sayisi

    @sayfa_sayisi.setter
    def sayfa_sayisi(self, deger):
        if deger <= 0:
            raise ValueError("Sayfa sayısı pozitif olmalıdır.")
        self._sayfa_sayisi = deger

    def __str__(self):
        return (f"[KİTAP] Kayıt No: {self._kayit_no} | "
                f"Başlık: {self._baslik} | "
                f"Yazar: {self._yazar} | "
                f"Sayfa: {self._sayfa_sayisi}")
class Dergi(Kaynak):
    def __init__(self, baslik, kayit_no, yayin_donemi, sayi_no):
        super().__init__(baslik, kayit_no)
        self._yayin_donemi = yayin_donemi
        self._sayi_no = sayi_no

    @property
    def yayin_donemi(self):
        return self._yayin_donemi

    @yayin_donemi.setter
    def yayin_donemi(self, deger):
        if not deger:
            raise ValueError("Yayın dönemi boş olamaz.")
        self._yayin_donemi = deger

    @property
    def sayi_no(self):
        return self._sayi_no

    @sayi_no.setter
    def sayi_no(self, deger):
        if deger <= 0:
            raise ValueError("Sayı numarası pozitif olmalıdır.")
        self._sayi_no = deger

    def __str__(self):
        return (f"[DERGİ]  Kayıt No: {self._kayit_no} | "
                f"Başlık: {self._baslik} | "
                f"Yayın Dönemi: {self._yayin_donemi} | "
                f"Sayı No: {self._sayi_no}")

class Islem(ABC):
    """CRUD işlemlerini zorunlu kılan soyut sınıf."""

    @abstractmethod
    def ekle(self):
        pass

    @abstractmethod
    def sil(self):
        pass

    @abstractmethod
    def guncelle(self):
        pass

    @abstractmethod
    def listele(self):
        pass

class KitapIslem(Islem):
    def __init__(self):
        self._kitaplar = []


    def kitap_sayisi(self):
        return len(self._kitaplar)


    def _kayit_no_ile_bul(self, kayit_no):
        for kitap in self._kitaplar:
            if kitap.kayit_no == kayit_no:
                return kitap
        return None

    def ekle(self):
        baslik    = input("Kitabın başlığını girin: ").strip()
        kayit_no  = input("Kitabın kayıt numarasını girin: ").strip()


        if self._kayit_no_ile_bul(kayit_no):
            print(f"[HATA] '{kayit_no}' numaralı kayıt zaten mevcut!\n")
            return

        yazar     = input("Kitabın yazarını girin: ").strip()
        try:
            sayfa = int(input("Kitabın sayfa sayısını girin: ").strip())
        except ValueError:
            print("[HATA] Sayfa sayısı tam sayı olmalıdır.\n")
            return

        kitap = Kitap(baslik, kayit_no, yazar, sayfa)
        self._kitaplar.append(kitap)
        print("Kitap başarıyla eklendi.")
        print(f"Toplam Kitap Sayısı: {self.kitap_sayisi()}\n")

    def sil(self):
        kayit_no = input("Silmek istediğiniz kitabın kayıt numarasını girin: ").strip()
        kitap = self._kayit_no_ile_bul(kayit_no)
        if kitap:
            self._kitaplar.remove(kitap)
            print(f"'{kitap.baslik}' adlı kitap silindi.\n")
        else:
            print(f"[HATA] '{kayit_no}' numaralı kitap bulunamadı.\n")

    def guncelle(self):
        kayit_no = input("Güncellemek istediğiniz kitabın kayıt numarasını girin: ").strip()
        kitap = self._kayit_no_ile_bul(kayit_no)
        if not kitap:
            print(f"[HATA] '{kayit_no}' numaralı kitap bulunamadı.\n")
            return

        print("Güncellemek istediğiniz alanı seçin:")
        print("  1) Başlık")
        print("  2) Yazar")
        print("  3) Sayfa Sayısı")
        secim = input("Seçiminiz: ").strip()

        if secim == "1":
            kitap.baslik = input("Yeni başlık: ").strip()
        elif secim == "2":
            kitap.yazar = input("Yeni yazar: ").strip()
        elif secim == "3":
            try:
                kitap.sayfa_sayisi = int(input("Yeni sayfa sayısı: ").strip())
            except ValueError:
                print("[HATA] Geçersiz değer.\n")
                return
        else:
            print("[HATA] Geçersiz seçim.\n")
            return

        print("Kitap başarıyla güncellendi.\n")

    def listele(self):
        if not self._kitaplar:
            print("Kayıt bulunamadı.\n")
            return
        print(f"\n── Kitap Listesi ({self.kitap_sayisi()} adet) ──")
        for kitap in self._kitaplar:
            print(kitap)
        print()

class DergiIslem(Islem):
    def __init__(self):
        self._dergiler = []

    def dergi_sayisi(self):
        return len(self._dergiler)

    def _kayit_no_ile_bul(self, kayit_no):
        for dergi in self._dergiler:
            if dergi.kayit_no == kayit_no:
                return dergi
        return None

    def ekle(self):
        baslik       = input("Derginin başlığını girin: ").strip()
        kayit_no     = input("Derginin kayıt numarasını girin: ").strip()


        if self._kayit_no_ile_bul(kayit_no):
            print(f"[HATA] '{kayit_no}' numaralı kayıt zaten mevcut!\n")
            return

        yayin_donemi = input("Yayın dönemini girin (Aylık/Haftalık vb.): ").strip()
        try:
            sayi_no = int(input("Sayı numarasını girin: ").strip())
        except ValueError:
            print("[HATA] Sayı numarası tam sayı olmalıdır.\n")
            return

        dergi = Dergi(baslik, kayit_no, yayin_donemi, sayi_no)
        self._dergiler.append(dergi)
        print("Dergi başarıyla eklendi.")
        print(f"Toplam Dergi Sayısı: {self.dergi_sayisi()}\n")

    def sil(self):
        kayit_no = input("Silmek istediğiniz derginin kayıt numarasını girin: ").strip()
        dergi = self._kayit_no_ile_bul(kayit_no)
        if dergi:
            self._dergiler.remove(dergi)
            print(f"'{dergi.baslik}' adlı dergi silindi.\n")
        else:
            print(f"[HATA] '{kayit_no}' numaralı dergi bulunamadı.\n")

    def guncelle(self):
        kayit_no = input("Güncellemek istediğiniz derginin kayıt numarasını girin: ").strip()
        dergi = self._kayit_no_ile_bul(kayit_no)
        if not dergi:
            print(f"[HATA] '{kayit_no}' numaralı dergi bulunamadı.\n")
            return

        print("Güncellemek istediğiniz alanı seçin:")
        print("  1) Başlık")
        print("  2) Yayın Dönemi")
        print("  3) Sayı No")
        secim = input("Seçiminiz: ").strip()

        if secim == "1":
            dergi.baslik = input("Yeni başlık: ").strip()
        elif secim == "2":
            dergi.yayin_donemi = input("Yeni yayın dönemi: ").strip()
        elif secim == "3":
            try:
                dergi.sayi_no = int(input("Yeni sayı numarası: ").strip())
            except ValueError:
                print("[HATA] Geçersiz değer.\n")
                return
        else:
            print("[HATA] Geçersiz seçim.\n")
            return

        print("Dergi başarıyla güncellendi.\n")

    def listele(self):
        if not self._dergiler:
            print("Kayıt bulunamadı.\n")
            return
        print(f"\n── Dergi Listesi ({self.dergi_sayisi()} adet) ──")
        for dergi in self._dergiler:
            print(dergi)
        print()

class Menu:
    def __init__(self):
        self._kitap_islem  = KitapIslem()
        self._dergi_islem  = DergiIslem()

    def goster(self):
        print("=" * 40)
        print("   KÜTÜPHANE YÖNETİM SİSTEMİ")
        print("=" * 40)
        print("1. Kitap Ekle")
        print("2. Kitap Sil")
        print("3. Kitap Güncelle")
        print("4. Kitapları Listele")
        print("5. Dergi Ekle")
        print("6. Dergi Sil")
        print("7. Dergi Güncelle")
        print("8. Dergileri Listele")
        print("9. Çıkış")
        print("-" * 40)

    def calistir(self):
        while True:
            self.goster()
            secim = input("Yapmak istediğiniz işlemi seçin (1-9): ").strip()

            if secim == "1":
                self._kitap_islem.ekle()
            elif secim == "2":
                self._kitap_islem.sil()
            elif secim == "3":
                self._kitap_islem.guncelle()
            elif secim == "4":
                self._kitap_islem.listele()
            elif secim == "5":
                self._dergi_islem.ekle()
            elif secim == "6":
                self._dergi_islem.sil()
            elif secim == "7":
                self._dergi_islem.guncelle()
            elif secim == "8":
                self._dergi_islem.listele()
            elif secim == "9":
                print("Programdan çıkılıyor.")
                break
            else:
                print("[HATA] Geçersiz seçim. Lütfen 1-9 arasında bir değer girin.\n")

if __name__ == "__main__":
    menu = Menu()
    menu.calistir()
```
